#### Tarea para la casa

Escribe un programa que solicite una puntuación entre 0.0 y 5.0. Si la puntuación está fuera de ese rango, muestra un mensaje de error. Si la puntuación está entre 0.0 y 5.0, muestra la calificación usando la tabla siguiente, para un número de estudiantes n:

|Puntuación|	Cualificación|
|----------|-----------|
|4.7 - 5.0|	Sobresaliente|
|4.0 - 4.7|	Notable|
|3.5 - 4.0|	Aceptable|
|3.0 - 3.5|	Suficiente|
|<3|Revisión|


```python
estudiantes = []
for i in range(3):
  nombre = input("Nombre: ")
  estudiantes.append(nombre)
```

```python
# Función que almacena la calificaión del estudiantes a partir de la nota obtenida en la lista llamada lisCua

def valor(dato):

  lisCua = []
  
  if dato < 3.0:
    valoracion="Revisión"  #Variable que se devuelve con la valoración del estudiante para almacenarse en el diccionario datEst
    lisCua.append("Revisión")
  elif dato >= 3.0 and dato < 3.5:
    valoracion="Suficiente"
    lisCua.append("Sufiente")
  elif dato >= 3.5 and dato < 4.0:
    valoracion="Aceptable"
    lisCua.append("Aceptable")
  elif dato >= 4.0 and dato < 4.7:
    valoracion="Notable"
    lisCua.append("Notable")
  elif dato >= 4.7 and dato <= 5.0:
    valoracion="Sobresaliente"
    lisCua.append("Sobresaliente")
  return valoracion
```

```python
# Función que almacena los estudiantes en la lista llamada lisEst

def estudiantes(nomb):
  lisEst = []
  lisEst.append(nomb)
  return
```

```python
# Diccionario para almacenar el nombre y la valoración del estudiantes

def datos(nomEst, nota):

  datEst = {}

  cualif = valor(nota)
  datEst.update({nomEst: cualif})
```

```python
numEst = int(input("¿Cuántos estudiantes son?"))
lisEst = []
lisCua = []
datEst = {}
seguir = "N"

for i in range(numEst):
  nomEst = input("Digite el nombre del Estudiante "+ str(i + 1) +": " )
  estudiantes(nomEst)
  while seguir == "N":
    nota = float(input("Digite la nota del Estudiante "+ nomEst + " "))
    if nota < 0.0 or nota > 5.0:
      print("Error, la nota debe estar en el rango entre 0.0 y 5.0")
    else:
      seguir = "S"

  valor(nota)
  datos(nomEst, nota)

for i in range(numEst):
  print("El Estudiante "+lisEst[i]+", obtuvo una cualificación "+lisCua[i])

# Print the value of datEst
#print(datEst)

# Complete the program validation:
# Terminate the program when the grade is less than zero (0) or greater than five (5)
```
